```python
import pandas as pd
import os
raw_data = pd.read_csv("/opt/ml/input/data/FE_total_data.csv")
raw_data['year'] = 0
raw_data['month'] = 0
raw_data['day'] = 0
raw_data.head()
```

```python
def year(s):
    y=s.split(' ')[0].split('-')[0]
    return int(y)

def month(s):
    m=s.split(' ')[0].split('-')[1]
    return int(m)

def day(s):
    d=s.split(' ')[0].split('-')[2]
    return int(d)

raw_data['year'] = raw_data["Timestamp"].apply(year)
raw_data['month'] = raw_data["Timestamp"].apply(month)
raw_data['day'] = raw_data["Timestamp"].apply(day)
raw_data.head()
```

```python
os.makedirs('/opt/ml/input/data/FE/', exist_ok=True)
raw_data.to_csv('/opt/ml/input/data/FE/yearMonthDay.csv', index=False)
```

```python
import pandas as pd
import os
from datetime import date
raw_data = pd.read_csv("/opt/ml/input/data/FE_total_data.csv")
raw_data['wday'] = 0
raw_data['weekNum'] = 0
raw_data['hour'] = 0
raw_data.head()
```

```python
def wday(time):
    return pd.Timestamp(time).dayofweek
def weekNum(s):
    y=int(s.split(' ')[0].split('-')[0])
    m=int(s.split(' ')[0].split('-')[1])
    d=int(s.split(' ')[0].split('-')[2])
    return date(y,m,d).isocalendar()[1]
def hour(s):
    h=s.split(' ')[1].split(':')[1]
    return int(h)


raw_data['wday'] = raw_data['Timestamp'].apply(wday)
raw_data['weekNum'] = raw_data["Timestamp"].apply(weekNum)
raw_data['hour'] = raw_data["Timestamp"].apply(hour)
raw_data.head()
```

```python
os.makedirs('/opt/ml/input/data/FE/', exist_ok=True)
raw_data.to_csv('/opt/ml/input/data/FE/wday.csv', columns=["userID", "assessmentItemID", "testId", "answerCode", "Timestamp", "KnowledgeTag", "dataset", "wday"], index=False)
raw_data.to_csv('/opt/ml/input/data/FE/weekNum.csv', columns=["userID", "assessmentItemID", "testId", "answerCode", "Timestamp", "KnowledgeTag", "dataset", "weekNum"], index=False)
raw_data.to_csv('/opt/ml/input/data/FE/hour.csv', columns=["userID", "assessmentItemID", "testId", "answerCode", "Timestamp", "KnowledgeTag", "dataset", "hour"], index=False)
```
